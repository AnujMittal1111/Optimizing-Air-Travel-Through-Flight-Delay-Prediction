Classification Model: Will the flight be delayed?
In this section, I built a model to predict whether a group of flights would have any delay of 15 minutes or more.

The original target column, arr_del15, contains the number of delayed flights. To convert this into a binary classification problem, I created a new column that marks a delay as 1 if any delay was recorded, and 0 otherwise.

To avoid data leakage, I carefully selected features that are known before delays happen. These include:

month: to capture seasonal effects

arr_flights: total number of scheduled flights in the group

carrier_encoded and airport_encoded: label-encoded versions of carrier and airport

Seasonal flags: is_winter, is_summer, and is_festive to capture likely busy periods

Since the dataset was highly imbalanced (far more "delay" rows than "no delay"), I used SMOTE to oversample the minority class during training.

I trained three models for comparison:

Random Forest 

Logistic Regression

XGBoost Classifier

Each was evaluated using precision, recall, F1-score.
